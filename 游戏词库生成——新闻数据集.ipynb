## Excel文件清洗，分词后的列表存为txt文件 

```python
import pandas as pd
data = pd.read_excel(r"D:\datasets\data test\新闻筛选后.xlsx").astype(str)          #读入处理后的新闻数据
data.columns = ['text','number']                                                    #dataframe列名称设置
data.head()                                                                         #展示dataframe的前五行
```

```python
from LAC import LAC                                                                  #导入百度LAC包
lac = LAC(mode='lac')                                                                #装载LAC模型
train_lac = [lac.run(line)[0] for line in data.text]                                 #运行分词结果，将分词后的文本保存到一个列表之中
```

### 列表转存为txt文件，文件清洗

```python
def text_save(filename, data):                                                          #filename为写入CSV文件的路径，data为要写入数据列表.
    file = open(filename,'w',encoding='utf8')                                           #‘a'写入 ,'w'覆盖写入       
    for i in range(len(data)):
        s = str(data[i]).replace('（','').replace('）','')                              #去除（）
        s = s.replace('[','').replace(']','')                                           #去除[],这两行按数据不同，可以选择
        s = s.replace('《','').replace('》','')                                         #去除书名号
        s = s.replace('“','').replace('”','')                                         #去除双引号
        s = s.replace(r"\u3000",'').replace(r'\ue40c','').replace('nan','')             #去除特殊字符
        s = s.replace('，','') .replace("'","")                                         #去除中文逗号、单引号
        s = s.replace(',',"").replace('、','')                                          #去除英文逗号、顿号
        s = s + '\n'                                                                    #每行末尾追加换行符
        file.write(s)
    file.close()
    print("保存分词文件成功")
filename = r'D:\datasets\data test\news_fenci_pure.txt'
text_save(filename, train_lac)
```

### 使用lac模型进行分词 

```python
df = pd.read_csv(r"D:\datasets\data test\news_fenci_pure.txt")                       #读入处理后的新闻数据
df.columns = ['text']                                                                #dataframe列名称设置
df.head()
```

```python
from LAC import LAC                                                                 #导入百度LAC包
lac = LAC(mode='seg')                                                                #装载LAC模型
part_lac = [lac.run(line) for line in df.text]                                      #运行分词结果，将分词后的文本保存到一个列表之中
```

## <font color = deepskyblue>一级检索</font>

- 使用word2vec进行词嵌入以及相似词库的查找

```python
from gensim.models import Word2Vec
model = Word2Vec(part_lac,vector_size=100,min_count=1)
```

<font color =red>运行大约3min</font>

```python
print('与游戏最接近的30个词语：')                          
search_one = [key for key in model.wv.similar_by_word('游戏',topn=30)]               #运用word2vec模型展示与游戏相似度最高的三十个词语
print(search_one)
most_related_list = [key[0] for key in model.wv.similar_by_word('游戏',topn=30)]     #游戏最相关的词语集合
print(most_related_list)                  
```

```python
game_similar_list = []                                                              #存放与游戏相关度最高的文本
game_similar_list = game_similar_list + most_related_list[:5]
game_similar_list
```

扩充1

```python
search_one = []
for i in range(5):
    a = [key[0] for key in model.wv.similar_by_word(game_similar_list[i],topn=30)]
    search_one = search_one + a
search_one
```

```python
len(search_one)
```

- 去重

```python
def de_duplication(list1):
    list2=[];
    for i in list1:
        if i not in list2:
            list2.append(i);
    return list2
search_one_u = de_duplication(search_one)
search_one_u
```

```python
len(search_one_u)
```

```python
game_similar_list = game_similar_list + search_one_u
game_similar_list
```

```python
search_two = []
for i in range(len(search_one_u)):
    b = [key[0] for key in model.wv.similar_by_word(search_one_u[i],topn=15)]
    search_two = search_two + b
search_two
```

```python
len(search_two)
```

- 去重

```python
search_two_u = de_duplication(search_two)
search_two_u
```
