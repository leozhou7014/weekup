- 函数val_num 计算dataframe一列中每个词出现的次数，返回id及出现次数

```python
from tqdm import tqdm
import time
def val_num(col):
    a = []           
    b = []
    for i in tqdm(range(len(col))):
        c = col[i]
        if c not in a:
            a.append(c)                 #列表a存放一列中的各值（不重复）
            b.append(1)                 #首次出现元素 次数为1
            index = len(b)-1            
        else:
            b[index] += 1               #再次出现的元素，对应次数加1
    return a,b                          #返回fan_id和其对应的次数
```

```python
s_t = time.time()
col = df_con_test['fan_id']             #fan_id中各个值出现的次数 即其关注kol的数目
print(val_num(col))
print(time.time()-s_t)
```

```python
list_user,list_num = val_num(col)
df_user_test = pd.DataFrame(data=list_user,columns=['user_id'])         #转换为dataframe
df_user_test
```

```python
df_user_test['focus_kol_num'] = pd.Series(list_num)         #添加列 关注kol数目
df_user_test
```
