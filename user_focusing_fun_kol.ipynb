- åŠ è½½fan-followæ•°æ®

```python
df_con1 = pd.merge(df_ff,df_kol,left_on='followed_id',right_on="kol_id")  
'''ç¬¬ä¸€æ¬¡mergeï¼Œå°†fan-followedçš„è¡¨æ ¼æ•°æ®ä¸å¤§ç±»kolçš„æ•°æ®è¿›è¡ŒåŒ¹é…ï¼Œå·¦é”®é€‰æ‹©è¢«å…³æ³¨å¯¹è±¡ï¼Œå³é”®
é€‰æ‹©kolçš„idï¼Œç›®çš„æ˜¯å°†å…³æ³¨å¯¹è±¡ä¸­æœ‰å¤§ç±»kolçš„å¯¹è±¡ç­›é€‰å‡ºæ¥ï¼Œåç»­å¾ˆå®¹æ˜“å¾—åˆ°è¿™äº›ç”¨æˆ·å…³æ³¨çš„kolçš„æ•°ç›®'''
df_con1
```

```python
len(df_con1['fan_id'].unique())                  #åœ¨2kwå…³æ³¨å…³ç³»ä¸­æœ‰æœ‰87wä¸ªå…³æ³¨æç¬‘ç±»åšä¸»ï¼Œå¯¹åº”10w+ç”¨æˆ·
```

```python
focus_fun = df_con1['fan_id'].value_counts()     #è®¡ç®—æ¯ä¸ªç”¨æˆ·å…³æ³¨çš„æç¬‘ç±»åšä¸»çš„æ•°ç›®
focus_fun
```

- å°†ç”¨æˆ·å…³æ³¨çš„æç¬‘ç±»kolçš„ç”¨æˆ·çš„æ•°ç›®ä¿¡æ¯è½¬æ¢ä¸ºdataframe

```python
df_ff_user = pd.DataFrame(focus_fun)
df_ff_user['user_id'] = df_ff_user.index
df_ff_user = df_ff_user.rename(columns={'fan_id':'num'})
df_ff_user = df_ff_user.reset_index(drop=True)
df_ff_user
```

- è¯»å–ç”¨æˆ·ä¿¡æ¯è¡¨æ ¼

```python
df_user_full = pd.read_csv('D:/datasets/large_file/user6m.csv',encoding='utf-8')
df_user_full = df_user_full.rename(columns={'_id':'user_id'})   #é‡å‘½ååˆ—å
df_user_full
```

```python
df_user_briinfo = df_user_full[['user_id','nick_name','authentication','brief_intro','fan_num',
           'follow_num','tweet_num']]             #é€‰å–ç”¨æˆ·ä¿¡æ¯è¡¨æ ¼ä¸­éƒ¨åˆ†åˆ—
```

```python
df_con2 = pd.merge(df_ff_user,df_user_briinfo,on='user_id')
'''ç¬¬äºŒæ¬¡mergeï¼Œå°†ç¬¬ä¸€æ¬¡åŒ¹é…å¥½çš„ç»“æœä¸ç”¨æˆ·ç®€è¦ä¿¡æ¯åŒ¹é…'''
'''é‡è¦å…³æ³¨ä¸¤ä¸ªå˜é‡ï¼šæ¯ä¸ªç”¨æˆ·å…³æ³¨çš„æç¬‘ç±»çš„KOLçš„æ•°ç›®ä»¥åŠç”¨æˆ·å…³æ³¨çš„æ€»äººæ•°'''
df_con2
```

- å‰”é™¤å¼‚å¸¸ç”¨æˆ·

```python
df_con2 = df_con2[-(df_con2['follow_num']==0)][-(df_con2['tweet_num']==0)][df_con2['num']<df_con2['follow_num']]
'å»é™¤å…³æ³¨æ•°ä¸º0æˆ–å¸–å­æ•°ä¸º0æˆ–è€…å…³æ³¨æ•°å°äºå…³æ³¨æç¬‘ç±»kolæ•°çš„å¼‚å¸¸ç”¨æˆ·'
df_con2
```

```python
df_con2['prop'] = df_con2['num']/df_con2['follow_num']   #æ¯ä¸ªç”¨æˆ·å…³æ³¨æç¬‘ç±»kolå æ‰€å…³æ³¨çš„äººæ•°çš„æ¯”ä¾‹
df_con2 = df_con2.reset_index(drop=True)
df_con2
```

```python
df_con2 = df_con2[df_con2['prop']<=0.5]                      #å°†æ¯”ä¾‹è¿‡å¤§çš„ç”¨æˆ·å‰”é™¤ï¼Œè¿™ç±»ç”¨æˆ·ä¸­å¼‚å¸¸çš„å¾ˆå¤š
df_con2.sort_values('prop',ascending=False,inplace=True)    #æŒ‰ç…§æ¯”ä¾‹è¿›è¡Œé™åºæ’åˆ—
df_con2 = df_con2.reset_index(drop=True)
df_con2
```

```python
df_con2 = df_con2[-df_con2['nick_name'].isna()]          #å»é™¤æ˜µç§°ä¸ºnahçš„ç”¨æˆ·
df_con2
```
