数据： 大类kol数据、fan-followed表格数据、用户信息数据

```python
import numpy as np
import pandas as pd
```

- 加载搞笑类kol

```python
df_kol = pd.read_csv('D:/datasets/qianyi/user_label/method/based_on_user_focusing_kol/file/kol/fun_gt10w_3.csv',encoding='utf-8')
df_kol
```

- 加载fan-follow数据

```python
df_ff = pd.read_csv('D:/datasets/large_file/fan_follow_2kw.csv',encoding='utf-8')
df_ff
```

```python
# len(df_ff['fan_id'].unique())
```

```python
df_con1 = pd.merge(df_ff,df_kol,left_on='followed_id',right_on="kol_id")  
'''第一次merge，将fan-followed的表格数据与大类kol的数据进行匹配，左键选择被关注对象，右键
选择kol的id，目的是将关注对象中有大类kol的对象筛选出来，后续很容易得到这些用户关注的kol的数目'''
df_con1
```

```python
len(df_con1['fan_id'].unique())                  #在2kw关注关系中有有87w个关注搞笑类博主，对应10w+用户
```

```python
focus_fun = df_con1['fan_id'].value_counts()     #计算每个用户关注的搞笑类博主的数目
focus_fun
```

- 将用户关注的搞笑类kol的用户的数目信息转换为dataframe

```python
df_ff_user = pd.DataFrame(focus_fun)
df_ff_user['user_id'] = df_ff_user.index
df_ff_user = df_ff_user.rename(columns={'fan_id':'num'})
df_ff_user = df_ff_user.reset_index(drop=True)
df_ff_user
```

- 读取用户信息表格

```python
df_user_full = pd.read_csv('D:/datasets/large_file/user6m.csv',encoding='utf-8')
df_user_full = df_user_full.rename(columns={'_id':'user_id'})   #重命名列名
df_user_full
```

```python
df_user_briinfo = df_user_full[['user_id','nick_name','authentication','brief_intro','fan_num',
           'follow_num','tweet_num']]             #选取用户信息表格中部分列
```

```python
df_con2 = pd.merge(df_ff_user,df_user_briinfo,on='user_id')
'''第二次merge，将第一次匹配好的结果与用户简要信息匹配'''
'''重要关注两个变量：每个用户关注的搞笑类的KOL的数目以及用户关注的总人数'''
df_con2
```

- 剔除异常用户

```python
df_con2 = df_con2[-(df_con2['follow_num']==0)][-(df_con2['tweet_num']==0)][df_con2['num']<df_con2['follow_num']]
'去除关注数为0或帖子数为0或者关注数小于关注搞笑类kol数的异常用户'
df_con2
```

```python
df_con2['prop'] = df_con2['num']/df_con2['follow_num']   #每个用户关注搞笑类kol占所关注的人数的比例
df_con2 = df_con2.reset_index(drop=True)
df_con2
```

```python
df_con2 = df_con2[df_con2['prop']<=0.5]                      #将比例过大的用户剔除，这类用户中异常的很多
df_con2.sort_values('prop',ascending=False,inplace=True)    #按照比例进行降序排列
df_con2 = df_con2.reset_index(drop=True)
df_con2
```

```python
df_con2 = df_con2[-df_con2['nick_name'].isna()]          #去除昵称为nah的用户
df_con2
```

### 去水

- 加载水军数据

```python
df_wa = pd.read_csv('D:/datasets/large_file/water_army_rule2.csv',encoding='utf-8')
df_wa
```

```python
df_wa_1 = df_wa[['_id','nick_name']]
df_con_wa = pd.merge(df_con2,df_wa_1,left_on='user_id',right_on='_id')
'''第三次merge，将用户关注kol的信息与水军信息进行匹配，下一步去除水军数据'''
df_con_wa
```

```python
# (df_con2['user_id'].isin(df_con_wa['user_id']))
```

```python
df_con_dwa = df_con2[-(df_con2['user_id'].isin(df_con_wa['user_id']))]  #去水军
df_con_dwa = df_con_dwa.reset_index(drop=True)
df_con_dwa
```

### 阈值设定

```python
df_ffu = df_con_dwa
```

```python
df_ffu = df_ffu[df_ffu['prop']>=0.01]         #筛选关注搞笑类博主比例大于等于0.01的用户
df_ffu = df_ffu.reset_index(drop=True)
df_ffu
```

```python
# df_ffu[df_ffu['num']>=5][df_ffu['prop']>=0.05]
```

```python
df_ffu = df_ffu[df_ffu['num']>=3]
df_ffu = df_ffu.reset_index(drop=True)
df_ffu
```

```python
df_ffu = df_ffu[df_ffu['num']>=5][df_ffu['prop']>=0.05]
df_ffu = df_ffu.reset_index(drop=True)
df_ffu
```
